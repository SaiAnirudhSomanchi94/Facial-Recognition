Encoding 128 Features and identifying Face Location of the images 

Training Dataset

```python
def encode_data(image_path):
#     print(image_path)
    img_1 = face_recognition.load_image_file(image_path)
    img_2 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
    faceloc = face_recognition.face_locations(img_1)
    img_name = str(image_path).split('//')[1]
    if len(faceloc) != 0:
        faceloc = faceloc[0]
        encoding = face_recognition.face_encodings(img_1)[0]
        main = encoding.tolist()
        main.append(img_name)
        person_details.append(main)
        print(faceloc)
        main_1 = list(faceloc)
        main_1.append(img_name)
        face_Location.append(main_1)
        print(len(person_details))
    else:
        Face_identification_failed.append(image_path)
    return 
```

```python
threads = ThreadPool(10)
results = threads.map(encode_data,person_path_list_train)
```

Test Data set

```python
def encode_data_test(image_path):
#     print(image_path)
    img_1 = face_recognition.load_image_file(image_path)
    img_2 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
    faceloc = face_recognition.face_locations(img_1)
    img_name = str(image_path).split('//')[1]
    if len(faceloc) != 0:
        faceloc = faceloc[0]
        encoding = face_recognition.face_encodings(img_1)[0]
        main = encoding.tolist()
        main.append(img_name)
        person_details_test.append(main)
        print(faceloc)
        main_1 = list(faceloc)
        main_1.append(img_name)
        face_Location_test.append(main_1)
        print(len(person_details_test))
    else:
        Face_identification_failed_test.append(image_path)
    return 
```

```python
results = threads.map(encode_data_test,person_path_list_test)
```

We want to identify how many PCA are actually required to explain the entire variance in data

```python
df = X_train.append(X_test)
df_y = y_train.append(y_test)
X = df[df.columns[0:]]
X.describe()
```

```python
X
```

```python
mean_vec = np.mean(X, axis=0)
cov_mat = (X - mean_vec).T.dot((X - mean_vec)) / (X.shape[0]-1)
print('Covariance matrix \n%s' %cov_mat)
```

```python
# eigendecomposition on covariance matrix
cov_mat = np.cov(X.T)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

Each Component explaining most variance in order

```python
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```

We can identify how many principal components are required in the below diagram which is around 70

```python
#Explained variance
pca = PCA().fit(X)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
pca.components_
```

```python
# For copying images in one directory
import os, shutil
def copytree(src, dst, symlinks=False, ignore=None):
    for item in os.listdir(src):
        s = os.path.join(src, item)
        d = os.path.join(dst, item)
        if os.path.isdir(s):
            shutil.copytree(s, d, symlinks, ignore)
        else:
            shutil.copy2(s, d)
```

Decision Tree
